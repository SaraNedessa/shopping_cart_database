```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
prod = pd.read_csv("products.csv")
sales = pd.read_csv("sales.csv")
ord = pd.read_csv("orders.csv")
cust = pd.read_csv("customers.csv")
```

```python
prod = prod.rename(columns={"product_ID": "product_id", "quantity": "quantity_available"})
prod.head()
```

```python
sales.head()
```

```python
ord.head()
```

```python
cust.head()
```

```python
# merge datasets on primary keys
data = sales.merge(ord, on = "order_id").merge(cust, on = "customer_id").merge(prod, on = "product_id")
data
```

```python
data.describe()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data['order_date'] = pd.to_datetime(data['order_date'])
data['delivery_date'] = pd.to_datetime(data['delivery_date'])

data['delivery_speed'] = data['delivery_date'] - data['order_date']

data['delivery_speed'] = data.delivery_speed.astype(str).str.split(' ').str[0]
data.delivery_speed
```

```python
total_price_sold = data.groupby("order_date").sum(numeric_only=True)[["total_price"]].reset_index()
total_price_sold
```

```python
# EDA
# line graphs to track prices, sales, orders over time

sns.lineplot (data = total_price_sold, x = "order_date", y = "total_price")
plt.xticks(rotation=45)
```

```python
sns.lineplot (data = data, x = "order_date", y = "delivery_speed")
plt.xticks(rotation=45)
```
